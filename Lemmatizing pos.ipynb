```python
import pickle
import pandas as pd
import nltk
from nltk.stem import WordNetLemmatizer
```

```python
jerry_pos_df = pd.read_pickle('pos/jerry_pos.csv')
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized = []
for j in range(45):
    temp = []
    for i in range(len(jerry_pos_df)):
        try:
            temp.append(lemmatizer.lemmatize(jerry_pos_df.iloc[i][j]))
        
        except:
            break
    lemmatized.append(temp)
lemmatized
```
